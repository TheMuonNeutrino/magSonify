## Limits to frequency space resolution
By testing with peaks at varying proximity, we can show that paulstretch using these settings can distingush frequencies of 200 Hz, but not 100 Hz difference. In practise, this level of distinction is not audible, as the peaks are very broad (and noisy), so there would be no resonances generated which would allow a listener to recognise a difference.

```python
stretch = 16
lowerDifference = 100
upperDifference = 200
freqs = np.array((200,200+lowerDifference,1000,1000+upperDifference))
baseMethods.plotPSDpaulstretchHarmonic(freqs,stretch,showPlot=False)
plt.xlim([0,3500])
plt.ylim([1e-4,1])
plt.show()
```

```python
freqs += 10000
baseMethods.plotPSDpaulstretchHarmonic(freqs,stretch,showPlot=False)
plt.xlim([9000,12500])
plt.ylim([1e-4,1])
plt.show()
```
